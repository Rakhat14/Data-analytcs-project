```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data
data = np.linspace(0, 9, 100)

# Plot the data
plt.plot(data, data, label='linear')
# Add label text
plt.ylabel('some numbers')
plt.xlabel('some numbers')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

<b>Task 2.1</b> Plotting bar charts
<br/>
On the example below, the bar chart for Seasons was plotted the documentation [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)

```python
plt.bar(data.Season.unique(), data.groupby('Season')['Sales'].sum(), width = 0.9, 
        color='#288664', ec='#288664')
plt.xlabel('Seasons')
plt.xticks(rotation = '45')
plt.ylabel('Sales amount')
plt.title('Sales over seasons')
plt.show()
```

<b>TASK 3</b> Scatterplots

Take a look at well known <u>Temperature - Sales relationship</u>, visualized by scatterplot


```python
plt.scatter(data.Temperature, data.Sales, 50, marker = '*', color = '#862844')
plt.yticks(np.arange(0, 60, 5))
plt.xticks(np.arange(10, 130, 10))
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.axis('equal')
plt.title('Sales and Temperature relationship')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(data.Sales, data.Temperature, color='green')
plt.xlabel('Temperature')
plt.xticks(rotation = '90')
plt.ylabel('Sales')
plt.title('Temperature + Sales ')
plt.show()

plt.subplot(1,2,2)
plt.hist( data.Temperature, color='b',histtype = 'step')
plt.xlabel('Temperature')
plt.xticks(rotation = '90')
plt.show()

plt.subplot(1,2,2)
plt.hist( data.Sales, color='b',histtype = 'step')
plt.xlabel('Sales')
plt.xticks(rotation = '90')
plt.show()
```
